```python
# f(x) = -12*x**4 - 18*x**3+5*x**2 + 10*x - 30
# 1. Определить корни
# 2. Найти интервалы, на которых функция возрастает
# 3. Найти интервалы, на которых функция убывает
# 4. Построить график
# 5. Вычислить вершину
# 6. Определить промежутки, на котором f > 0
# 7. Определить промежутки, на котором f < 0
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sympy import *
```

```python
limit = 2 # устанавливаем лимит для границ (откуда рисуем)
step = 0.01 # шаг
color = 'b'
line_s = '-'
direct_up = True # маркер направления (up - down)
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
```

```python
x = np.arange(-limit, limit, step)
```

```python
def switch_color(): # для смены цвета - убывание - возрастание
    global color
    if color == 'b':
        color = 'r'
    else:
        color = 'b'
    return color
```

```python
def switch_line(): # для отображения f > 0 или f < 0
    global line_s
    if line_s == '-':
        line_s = '--'
    else:
        line_s = '-'
    return line_s
```

```python
def func(x):
    return a * x**4 + b *x**3 + c *x **2 + d * x + e
```

```python
plt.rcParams['lines.linestyle'] = '-'
x_lim = np.arange(-1000, 1000, step)
plt.plot(x_lim, func(x_lim), color)
plt.grid()
```

```python
x_change = [(-limit, 'limit')] # заполняем x_change - -limit с комментарием 'limit' - от этой точки начинаем строить график

for i in range(len(x) - 1): # идем по длине x = np.arange(-limit, limit, step)
    if (func(x[i]) > 0 and func(x[i + 1]) < 0) or (func(x[i]) < 0 and func(x[i + 1]) > 0):
        x_change.append((x[i] if abs(0 - x[i]) < abs(0 - x[i+1]) else x[i+1], 'zero')) # маркер отображения пересечения
   # присвоим x[i] нулю если x[i] ближе к нулю иначе присваиваем x[i+1]
    if direct_up:
        if func(x[i]) > func(x[i + 1]):
            x_change.append((x[i], 'direct')) # маркер отображения смены направления
            direct_up = False
    else:
        if func(x[i]) < func(x[i + 1]):
            x_change.append((x[i], 'direct')) # маркер отображения смены направления
            direct_up = True

x_change.append((limit, 'limit'))       
print(x_change)              
```

```python
x_change.append((limit, 'limit'))
```

```python
# находим вершину (экстремум)
max_y = max(func(x))
max_x = -limit
for x in x_change:
    if x[1] in ['direct','limit']: # идем по 'перегибам' функции
        if abs(func(x[0]) - max_y) < abs(max_x - max_y):
            max_x = x[0]
```

```python
for i in range(len(x_change) - 1): # идем по длине x_change
    cur_x = np.arange(x_change[i][0], x_change[i + 1][0] + step, step) # включая точку x_change[i + 1][0] (поэтому + step)
    # (-10, 'limit'), (-9.979000000000049, 'direct')  -> (x_change[i][0] обращаемся к нуливой позиции кортежа (х+коммент)
    if x_change[i][1] == 'zero':
        plt.rcParams['lines.linestyle'] = switch_line() # меняем линию при пересечении оси х
        plt.plot(cur_x, func(cur_x),color)
    else:
        plt.plot(cur_x, func(cur_x), switch_color()) # меняем цвет при смене направления



plt.plot(max_x, max_y, 'yo', label=f'Экстремум функции на промежутке [{-limit};{limit}]: ({round(max_x, 2)}, {round(max_y, 2)})')

plt.rcParams['lines.linestyle'] = '-'
plt.plot(0, 0, 'b', label='Убывание < 0')
plt.plot(0, 0, 'r', label='Возрастание < 0')
plt.rcParams['lines.linestyle'] = '--'
plt.plot(0, 0, 'b', label='Убывание > 0')
plt.plot(0, 0, 'r', label='Возрастание > 0')
plt.legend()
plt.grid()
plt.show()
```
